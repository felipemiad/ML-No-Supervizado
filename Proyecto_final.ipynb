```python
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler

# Tomar un subconjunto aleatorio del 10% de los datos
df_sample = df[['Promedio_Dias_Hasta_Siniestro', 'Total_Desembolso']].sample(frac=0.1, random_state=42)

# Normalizar los datos
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df_sample)

# K-Means
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)  # Reduce n_init si es necesario
kmeans_labels = kmeans.fit_predict(data_scaled)
silhouette_kmeans = silhouette_score(data_scaled, kmeans_labels)
print(f'Silhouette Score para K-Means: {silhouette_kmeans}')

# DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(data_scaled)


```

```python
silhouette_dbscan = silhouette_score(data_scaled, dbscan_labels) if len(set(dbscan_labels)) > 1 else 'No se puede calcular'
print(f'Silhouette Score para DBSCAN: {silhouette_dbscan}')
```

```python
# Graficar los clusters de K-Means
plt.figure(figsize=(14, 6))

# Subgráfico para K-Means
plt.subplot(1, 2, 1)
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=kmeans_labels, cmap='viridis', marker='o', edgecolor='k', s=50)
plt.title('Clusters K-Means')
plt.xlabel('Promedio_Dias_Hasta_Siniestro (escala)')
plt.ylabel('Total_Desembolso (escala)')
plt.colorbar(label='Etiqueta del cluster')

# Subgráfico para DBSCAN
plt.subplot(1, 2, 2)
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=dbscan_labels, cmap='plasma', marker='o', edgecolor='k', s=50)
plt.title('Clusters DBSCAN')
plt.xlabel('Promedio_Dias_Hasta_Siniestro (escala)')
plt.ylabel('Total_Desembolso (escala)')
plt.colorbar(label='Etiqueta del cluster')

plt.tight_layout()
plt.show()
```

```python
# Agregar las etiquetas de los clusters al DataFrame
df_sample['Cluster_KMeans'] = kmeans_labels
df_sample['Cluster_DBSCAN'] = dbscan_labels

# Conteo de clusters y características para K-Means
num_clusters_kmeans = len(set(df_sample['Cluster_KMeans'])) - (1 if -1 in df_sample['Cluster_KMeans'] else 0)
descripcion_clusters_kmeans = df_sample.groupby('Cluster_KMeans')[['Promedio_Dias_Hasta_Siniestro', 'Total_Desembolso']].mean()

# Conteo de clusters y características para DBSCAN
num_clusters_dbscan = len(set(df_sample['Cluster_DBSCAN'])) - (1 if -1 in df_sample['Cluster_DBSCAN'] else 0)
descripcion_clusters_dbscan = df_sample.groupby('Cluster_DBSCAN')[['Promedio_Dias_Hasta_Siniestro', 'Total_Desembolso']].mean()

# Resultados
print(f'Número de clusters para K-Means: {num_clusters_kmeans}')
print('Descripción de clusters K-Means:')
print(descripcion_clusters_kmeans)

print(f'Número de clusters para DBSCAN: {num_clusters_dbscan}')
print('Descripción de clusters DBSCAN:')
print(descripcion_clusters_dbscan)

```

```python
import pandas as pd
import gower
from sklearn.cluster import DBSCAN
import numpy as np
# Tomar una muestra de los datos
df_sample = df.sample(n=5000, random_state=42)
# Convertir 'Nombre_Intermediario' a una representación numérica
df_sample['Nombre_Intermediario_Cod'] = df_sample['Nombre_Intermediario'].astype('category').cat.codes

# Asegurarse de que 'Promedio_Dias_Hasta_Siniestro' esté en formato flotante
df_sample['Promedio_Dias_Hasta_Siniestro'] = df_sample['Promedio_Dias_Hasta_Siniestro'].astype(float)

# Seleccionar las variables para clustering
variables_clustering = df_sample[['Promedio_Dias_Hasta_Siniestro', 'Nombre_Intermediario_Cod']]

# Calcular la matriz de distancia de Gower
gower_dist_matrix = gower.gower_matrix(variables_clustering)

# DBSCAN
dbscan = DBSCAN(eps=0.01, min_samples=5, metric='precomputed')
dbscan_labels = dbscan.fit_predict(gower_dist_matrix)

# Agregar etiquetas de clusters al DataFrame original
df_sample['Cluster_DBSCAN'] = dbscan_labels

# Ver los resultados
cluster_counts = pd.Series(dbscan_labels).value_counts()
print('Conteo de clusters DBSCAN:')
print(cluster_counts)

# Descripción de los clusters
descripcion_clusters_dbscan = df_sample.groupby('Cluster_DBSCAN')[['Promedio_Dias_Hasta_Siniestro']].mean()
print('Descripción de clusters DBSCAN:')
print(descripcion_clusters_dbscan)

# Ver los intermediarios en cada cluster
intermediarios_por_cluster = df_sample.groupby('Cluster_DBSCAN')['Nombre_Intermediario'].unique().reset_index()
intermediarios_por_cluster.columns = ['Cluster', 'Intermediarios']
print(intermediarios_por_cluster)

```

```python
import matplotlib.pyplot as plt

import matplotlib.pyplot as plt

# Definir la función para graficar el número de ocurrencias de intermediarios por cluster
def graficar_cluster(resultado, cluster_num):

    # Ordenar los valores del cluster seleccionado
    valores_ordenados = resultado[cluster_num].sort_values(ascending=False).head(10)

    # Crear el gráfico de barras
    plt.figure(figsize=(10, 6))
    plt.bar(valores_ordenados.index, valores_ordenados.values)

    # Ajustar las etiquetas del eje x
    plt.xticks(rotation=90)

    # Agregar título y etiquetas de ejes
    plt.title(f'Intermediarios financieros - Cluster KMeans {cluster_num}', fontsize=14)
    plt.xlabel('Nombre del Intermediario', fontsize=12)
    plt.ylabel('Número de ocurrencias', fontsize=12)

    # Mostrar el gráfico
    plt.tight_layout()
    plt.show()
graficar_cluster(resultado,0)

```

```python
import matplotlib.pyplot as plt

# Crear la figura con subplots 3x2
fig, axes = plt.subplots(2, 3, figsize=(17, 15))
axes = axes.flatten()  # Aplanar los ejes para fácil acceso

# Lista de clusters que vamos a graficar
clusters_a_graficar = [0, 1, 2, 3, 4]

# Iterar sobre los clusters y los ejes
for i, cluster_num in enumerate(clusters_a_graficar):
    # Seleccionar el eje actual donde graficar
    plt.sca(axes[i])  # Establecer el subplot en el que se graficará
    # Llamar a la función que ya creaste para graficar el cluster
    valores_ordenados = resultado[cluster_num].sort_values(ascending=False).head(10)

    # Crear el gráfico de barras en el subplot correspondiente
    axes[i].bar(valores_ordenados.index, valores_ordenados.values)

    # Ajustar las etiquetas del eje x
    axes[i].set_xticklabels(valores_ordenados.index, rotation=90)

    # Agregar título y etiquetas de ejes
    axes[i].set_title(f'Intermediarios financieros - Cluster KMeans {cluster_num}', fontsize=14)
    axes[i].set_xlabel('Nombre del Intermediario', fontsize=12)
    axes[i].set_ylabel('Número de ocurrencias', fontsize=12)

# Eliminar el último subplot vacío si no hay un gráfico
fig.delaxes(axes[-1])  # Eliminamos la sexta posición en la grilla si está vacía.

# Ajustar el layout para evitar superposiciones
plt.tight_layout()
plt.show()

```
